# Internet Users Data Visualization with Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import data
stats = pd.read_csv('DemographicData.csv')
```

```python
stats.columns = ['CountryName','CountryCode','BirthRate','InternetUsers','IncomeGroup']
stats.head()
```

# Distribution Plot

```python
# Distribution
plt.rcParams["patch.force_edgecolor"] = True
plt.rcParams['figure.figsize'] = 8,4
vis1 = sns.distplot(stats['InternetUsers'])
vis1
```

```python
vis1 = sns.distplot(stats['InternetUsers'], bins = 30)
```

We can see there is a cluster around 0 to 20% and begins to drop off around 75%.  
In other words, the majority of countries are sitting somewhere under 75% of their population having access to the Internet.

# Boxplots

```python
vis2 = sns.boxplot(data = stats, x = 'IncomeGroup', y = 'BirthRate')
```

# Scatterplots

```python
vis3= sns.lmplot(data = stats, x = 'InternetUsers', y = 'BirthRate',
                 fit_reg= False, hue = 'IncomeGroup', size = 10, aspect =1) #Plot data and regression model fits across a FacetGrid so have to specify size separately
```

We can see the that countries with a higher percentage of Internet users tend to have a lower birth rate.

# Keyword Arguments in Python

```python
# Marker Size
vis3 = sns.lmplot(data= stats, x = 'InternetUsers', y = 'BirthRate',
                 fit_reg = False, hue = 'IncomeGroup', size = 20, #increases facet size
                 scatter_kws = {'s':200}) # increaes marker size
```
