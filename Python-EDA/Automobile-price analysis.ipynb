```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
loc_path = 'C:/Users/132004/Downloads/datasets/attachment_Project_4_Automobile_data/Project_4_Automobile data'
```

```python
automob = pd.read_csv(loc_path + '/Automobile_data.txt',na_values='?')
pd.set_option('display.max_columns', 500)
automob.head()
```

Exploring dataset for null values 

```python
automob.info()                          #So, most of the null values are there in the 'normalized-losses' column
```

```python
automob.isnull().sum()              #returns no. of NaN values in each column
```

```python
#Filling normalized-losses NaN values by mean & rest of the columns by backfill method
automob['normalized-losses'] = automob['normalized-losses'].fillna(value=automob['normalized-losses'].mean())
automob = automob.fillna(method='bfill')
automob.head()
```

# Bivariate Analysis: Analysis to observe different parameters' impact on the pricing

```python
automob.groupby('num-of-doors')['price'].agg({'mean':np.mean,'max':np.max})   #Not much impact
```

```python
automob.groupby('aspiration')['price'].agg({'mean':np.mean,'max':np.max})   #As expected, turbo automobiles are bit 
                                                                            #expensive
```

```python
automob.groupby('fuel-type')['price'].agg({'mean':np.mean,'max':np.max})  #Diesel- based automobiles are bit expensive
#                                                                          than gas- based.
```

```python
automob.groupby('make')['price'].agg({'mean':np.mean,'max':np.max}).sort_values(by='mean',ascending=False)

#So, for few 'make', price is significantly higher than the rest.(Observe the drop for the mean price value from 
#                                                                 "BMW" to "Volvo")
```

## Using plots for the Bivariate analysis

```python
#Plotting cataegorical vs numerical data(Make v/s price) using swarmplot

import seaborn as sns
import matplotlib.pyplot as plt
sns.swarmplot(x="make", y="price", data=automob);
plt.xticks(rotation=90)
```

```python
#Using boxplot to gain more statistical insight for the Make v/s price analysis.

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7)) 
sns.boxplot(x="make", y="price", data=automob);
plt.xticks(rotation=90)
```

### Observations from the boxplot:
    1. The most-expensive-cars are: Mercedes-Benz, BMW, Porcshe & Jaguar with price value mostly above 20000 mark, 
       BMW being the only one below 20000 price mark for few cases.
    2. Automobiles for other makes are mostly below 20000 price mark, Volvo being the only one valued above 20000 
       price mark for few cases.

```python
#Plotting correlation between all the numerical parameters for the automobiles.

plt.figure(figsize=(15,7))
sns.heatmap(automob.corr(),annot=True)
```

### Observation from the heatmap:
    Price is mostly affected by engine-size, curb-weight, horsepower, width, length & mpg (inverse relation) of 
    automobile.

```python
# Plotting and validating the relation b/w price and highly significant parameters

plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.scatter('engine-size','price',data = automob)
plt.title("Engine-size v/s Price")
plt.subplot(2,2,2)
plt.scatter('curb-weight','price',data = automob)
plt.title("Curb-weight v/s Price")
plt.subplot(2,2,3)
plt.scatter('width','price',data = automob)
plt.title("Width v/s Price")
plt.subplot(2,2,4)
plt.scatter('city-mpg','price',data = automob)
plt.title("City-mpg v/s Price")
```

### Observation from the scatter plots:
    1. Curb-weight and price relation is not significant for higher values of curb-weight.
    2. City-mpg doesn't have significant effect on the price for values less than 22.
    3. Width and price relation is not significant for higher values of width(70-72).

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot('drive-wheels','price',data = automob)
plt.subplot(2,2,2)
sns.boxplot('body-style','price',data = automob)
plt.subplot(2,2,3)
sns.boxplot('num-of-cylinders','price',data = automob)
plt.subplot(2,2,4)
sns.boxplot('symboling','price',data = automob)
```

### Observations from the boxplots:
    1. Automobiles with Rear wheel drive are way more expensive than remaining automobiles.
    2. Price's IQR for automobiles with 'Hardtop' body-style is highest.
    3. Automobiles with 'hatchback' & 'wagon' body-style is much costlier than other automobiles.
    4. Price's IQR for automobiles with symboling '-1' is highest.
    5. Automobiles with symboling '-1' and '0' have higher prices than rest.
    6. As no. of cylinders increases, price of the automobile also increases; automobiles with eight cylinders being
       the most expensive ones.
